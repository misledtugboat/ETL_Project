```python
import pandas as pd
```

```python
movie_genres_df = pd.read_csv('Resources/movie_genres.csv')

movie_genres_df.head()
```

```python
movie_genres_df['year'] = movie_genres_df['title'].str[-5:]
movie_genres_df['year'] = movie_genres_df['year'].str[:-1]
movie_genres_df['title'] = movie_genres_df['title'].str[:-6]

movie_genres_df.head()
```

```python
movie_genres_df = movie_genres_df[~movie_genres_df['genres'].isin(['(no genres listed)'])]

movie_genres_df[['genre','genre_1','genre_2','genre_3','genre_4','genre_5','genre_6','genre_7',
                 'genre_8','genre_9']] = movie_genres_df['genres'].str.split('|',expand=True)

movie_genres_df.head()
```

```python
movie_genres_df = movie_genres_df[['title','year','genre']]
movie_genres_df.dropna()

movie_genres_df['title'] = movie_genres_df['title'].str.lower()
movie_genres_df['genre'] = movie_genres_df['genre'].str.lower()

movie_genres_df.head()
```

```python
movie_genres_df['title'].str.strip()
movie_genres_df['year'].str.strip()
movie_genres_df['genre'].str.strip()

movie_genres_df = movie_genres_df[['title','genre']]
movie_genres_df
```
